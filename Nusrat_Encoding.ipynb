**One Hot Encoding**

```python
df4.head()
```

```python
dummies = pd.get_dummies(df4[['sex', 'smoker', 'region']], drop_first=True, prefix=['sex', 'smoker', 'region'])
dummies.head()
```

```python
df4 = df4.drop(['sex', 'smoker', 'region'], axis=1)
df4.head()
```

```python
df4 = pd.concat([df4, dummies], axis=1)
df4.head()
```

```python
df5.head()
```

```python
# Loop

for col in df5.columns:
    if df5[col].dtypes == "object":
        oeh = pd.get_dummies(df5[col], drop_first=True, prefix=col)
        df5 = df5.drop(df5[[col]], axis=1)
        df5 = pd.concat([df5, oeh], axis=1)
```

**Ordinal**

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df6.head()
```

```python
unique_data = [df6.sex.unique(), df6.smoker.unique(), df6.region.unique()]
unique_data
```

```python
oe = OrdinalEncoder(categories=unique_data)
```

```python
encode = oe.fit_transform(df6[['sex','smoker','region']])
```

```python
# Loop
df7.head()
```

```python
for col in df7.columns:
    if df7[col].dtypes == 'object':
        unique_data2 = df7[col].unique()
        oe2 = OrdinalEncoder(categories=[unique_data2])
        encode2 = oe2.fit_transform(df7[[col]])
        df7[col] = encode2
```
